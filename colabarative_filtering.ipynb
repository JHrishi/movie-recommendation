```python
#imports
import pandas as pd
from scipy import sparse
```

```python
#https://github.com/babu-thomas/movielens-posters
# https://grouplens.org/datasets/movielens/100k/
ratings = pd.read_csv('info.csv')
movies = pd.read_csv('movies_data.csv')
movie_posters = pd.read_csv('movie_poster.csv',names=['item_id','poster_url'])
movie_url = pd.read_csv('movie_url.csv',names=['item_id','url'])
```

```python
movies = movies.merge(movie_posters,on='item_id')
movies = movies.merge(movie_url,on='item_id')
```

```python
movies
```

```python
ratings = ratings.merge(movies,on='item_id').drop(['timestamp'],axis=1)
print(ratings.shape)
ratings.head()
```

```python
ratings['title'].nunique()
```

```python
userRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
userRatings.head()
print("Before: ",userRatings.shape)
userRatings = userRatings.dropna(thresh=10, axis=1).fillna(0,axis=1)
#userRatings.fillna(0, inplace=True)
print("After: ",userRatings.shape)
```

```python
userRatings
```

```python
corrMatrix = userRatings.corr(method='pearson')
corrMatrix.head(100)
```

```python
def get_similar(movie_name,rating):
    similar_ratings = corrMatrix[movie_name]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_ratings
```

```python
similar_movies = get_similar("12 Angry Men (1957)",4)
```
